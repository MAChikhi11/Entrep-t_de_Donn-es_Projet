```python
#Joindre plusieurs fichiers csv en un seul
import pandas as pd
import glob
import os
from sklearn.impute import SimpleImputer
```

```python
# merging the files
files_joined = os.path.join(
    "./Dataset/Weather Data/Algeria", "Weather_*.csv"
)
```

```python
# Return a list of all joined files
list_files = glob.glob(files_joined)
```

```python
print("** Merging multiple csv files into a single pandas df1 **")
# Merge files by joining all files
df1 = pd.concat(map(pd.read_csv, list_files), ignore_index=True)
print(df1.head())
print(df1.isnull().sum())
```

```python
# Save the merged file
df1.to_csv("./Dataset/Weather Data/Algeria/Weather_Algeria.csv", index=False)
```

```python
# set files path
algeria = "./Dataset/Weather Data/Algeria/Weather_Algeria.csv"
morocco = "./Dataset/Weather Data/Morocco/Weather_Morocco.csv"
tunisia = "./Dataset/Weather Data/Tunisia/Weather_Tunisia.csv"
```

```python
print("*** Merging multiple csv files into a single pandas dataframe ***")

# merge files
dataFrame = pd.concat(map(pd.read_csv, [algeria, morocco, tunisia]), ignore_index=True)
print(dataFrame.head())
```

```python
keep_columns = ['STATION', 'NAME', 'LATITUDE', 'LONGITUDE', 'ELEVATION', 'DATE', 'PRCP', 'TAVG', 
                'TMAX', 'TMIN', 'SNWD', 'PGTM', 'SNOW', 'WDFG', 'WSFG',]

dataFrame.drop(columns=[col for col in dataFrame.columns if col not in keep_columns], inplace=True)
```

```python
# Convert date to datetime format
dataFrame['DATE'] = pd.to_datetime(dataFrame['DATE'], format='%Y-%m-%d')

# Create new columns for year, month, and day
dataFrame['YEAR'] = dataFrame['DATE'].dt.year
dataFrame['MONTH'] = dataFrame['DATE'].dt.month
dataFrame['DAY'] = dataFrame['DATE'].dt.day
```

```python
dataFrame['PAYS'] = dataFrame['STATION'].str[:2]
```

```python
dataFrame['TMIN'] = dataFrame['TMIN'].replace(-99, pd.NA)
dataFrame['TMIN'] = dataFrame['TMIN'].replace(93.2, pd.NA)
```

```python
dataFrame['PRCP'].fillna(0, inplace=True)
dataFrame['SNWD'].fillna(0, inplace=True)
dataFrame['SNOW'].fillna(0, inplace=True)

dataFrame['TAVG'].fillna(dataFrame['TAVG'].median(), inplace=True)
dataFrame['TMAX'].fillna(dataFrame['TMAX'].median(), inplace=True)
dataFrame['TMIN'].fillna(dataFrame['TMIN'].median(), inplace=True)
dataFrame['PGTM'].fillna(dataFrame['PGTM'].median(), inplace=True)
dataFrame['WDFG'].fillna(dataFrame['WDFG'].median(), inplace=True)
dataFrame['WSFG'].fillna(dataFrame['WSFG'].median(), inplace=True)
```

```python
dataFrame.to_csv("./Dataset/Weather_data.csv", index=False)
```
